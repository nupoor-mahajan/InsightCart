```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.dropna(subset=['Review Text', 'Rating'], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Rating'].head(10)
```

```python
df['Rating'] = df['Rating'].astype(str).str.extract('(\d+)')
```

```python
df['Rating'] = pd.to_numeric(df['Rating'])
```

```python
df['Rating'].head(10)
```

```python
def sentiment_label(rating):

    if rating <= 2:
        return "Negative"

    elif rating == 3:
        return "Neutral"

    else:
        return "Positive"
```

```python
df['Sentiment'] = df['Rating'].apply(sentiment_label)
```

```python
df[['Rating', 'Sentiment']].head()
```

```python
df['Sentiment'].value_counts()
```

```python
df = df[df['Reviewer Name'] != 'Reviewer Name']
```

```python
df.reset_index(drop=True, inplace=True)
```
